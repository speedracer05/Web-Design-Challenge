```python
import pandas as pd
import os
```

```python
# file = pandas.read_csv("./Resources/cities.csv", encoding="utf-8", nrows=10)
csv_path = os.path.join('Resources', 'cities.csv')
cities_csv = pd.read_csv(csv_path, encoding="utf-8")
cities_csv
```

```python
# drop the rows containing special characters
cities_drop = cities_csv.drop(cities_csv[cities_csv.City.str.contains(r'[^0-9a-zA-Z]')].index)
cities_drop.to_csv('Resources/cities_drop.csv')
```

```python
cities_drop.to_html('table.html', index=False, classes=['table', 'table-striped', 'table-hover'])
```
